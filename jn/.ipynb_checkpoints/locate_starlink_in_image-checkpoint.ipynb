```python
# select all starlinks in passages file for each day (i.e. first + second day of diff of 2 images)

def starlinks_in_diffimg(image):
    
#     img = glob.glob(f'../test_data/diff_{image}{camera}.fits.gz')
    img = f'../test_data/diff_{image}LSC.fits.gz'
    _, header = pf.getdata(img, header=True) 
        
    jd0 = header[7]
    jd1 = header[8]
        
    jd0_starlinks = []
    for sat in starlinks_no:
        i = np.where(passages[jd0] == sat)[0]
        if i.size > 0:
            jd0_starlinks.append(sat)

    jd1_starlinks = []
    for sat in starlinks_no:
        i = np.where(passages[jd1] == sat)[0]
        if i.size > 0:
            jd1_starlinks.append(sat)
            
    return jd0_starlinks, jd1_starlinks
```

```python
jd0_starlinks, jd1_starlinks = starlinks_in_diffimg(image=48510733)
```
