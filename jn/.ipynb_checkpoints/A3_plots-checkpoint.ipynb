```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

path = '/Users/peter/Projects/RL/A3/beta_tuning/'
```

```python
def get_rewards(beta, filter_size=201):
    
    rewards = []
    min_length = np.inf
        
    for i in range(5):
        score = np.load(f'{path}score_{beta}_{i}.npy')
        rewards.append(score)
        min_length = min(min_length, len(score))
    
    rewards = np.array([r[:min_length] for r in rewards])
    smoothed = savgol_filter(rewards, filter_size, 1)
    mean = np.mean(smoothed, axis=0)
    std = np.std(smoothed, axis=0)
    
    return mean[:min_length], std[:min_length], min_length


plt.figure(figsize=[8,6])
beta = ['0.1', '0.075']
for i in range(len(beta)):
    R, std, min_length = get_rewards(beta[i])
    #plt.fill_between(np.arange(len(R)), R-std, R+std, color=f'C{i}', alpha=0.5)
    plt.plot(np.arange(len(R)), R, color=f'C{i}', label=beta[i])

plt.title('Beta', fontsize=14)
plt.ylabel('Reward', fontsize=14)
plt.xlabel('Episode', fontsize=14)
plt.legend(loc=2, fontsize=14, frameon=False)
plt.xlim(0, min_length)
plt.tight_layout()
# plt.savefig('beta.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3, figsize=[10,6])

for i, ax in enumerate([ax1,ax2,ax3,ax4,ax5]):
    score = np.load(f'{path}score_0.1_{i}.npy')
    r = savgol_filter(score, 201, 1)
    ax.plot(np.arange(len(r)), r, c='C0')

ax6.axis('off')
text_kwargs = dict(ha='center', va='center', fontsize=20, color='k')
ax6.text(0.5, 0.5, 'Beta = 0.1', **text_kwargs)
plt.tight_layout()
# plt.savefig('beta_0.1.png', bbox_inches='tight', dpi=300)
plt.show()
```
